```python
DIRECT = True
UNDIRECT = True
consonants = "б в г д ж з к л м н п р с т ф х ц ч ш щ"
vowels = "а о е і и у я ю"
consonants = consonants.split(' ')
vowels = vowels.split(' ')
syllables1 = []
syllables2 = []
if DIRECT:
    for c in consonants:
        for v in vowels:
            syllables1.append(c + v)
if UNDIRECT:
    for c in consonants:
        for v in vowels:
            if v != 'ьо':
                syllables2.append(v + c)
            
syllables = syllables1 + syllables2
print(syllables)
```

```python
import random
import time
random.seed(time.time())
random.shuffle(syllables)
print(syllables)
```

```python
from sudoku import Sudoku
SIZE = 2
DIFFICULTY = 0.5
COL = 3
ROW = 4
PAGES = 1
NUM = COL * ROW * PAGES
start_idx = 0
puzzles = []
for i in range(NUM):
    end_idx = start_idx + SIZE * SIZE
    if end_idx > len(syllables):
        start_idx = 0
        random.shuffle(syllables)
    selected = syllables[start_idx:end_idx]
    start_idx = start_idx + end_idx
    
    puzzle = Sudoku(SIZE, seed=time.time()).difficulty(DIFFICULTY)
    res = ""
    for line in puzzle.board:
        for e in line:
            res += str(e) + '\t'
        res += '\n'
    res = res[:-1]
    keys = list(range(1, SIZE * SIZE + 1))
    values = selected
    d = dict(map(lambda i,j : (i,j) , keys,values))
    d["None"] = ""
    for e in d:
        res = res.replace(str(e), d[e])
    puzzles.append(res)
    print(res)
```
